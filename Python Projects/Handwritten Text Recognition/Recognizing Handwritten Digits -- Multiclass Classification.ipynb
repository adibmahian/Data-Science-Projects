# Recognizing Handwritten Digits - Multiclass Classification 

```python
from sklearn.datasets import load_digits
%matplotlib inline
import matplotlib.pyplot as plt
digits = load_digits()
```

```python
plt.gray() 
for i in range(10):
    plt.matshow(digits.images[i]) 
```

```python
digits.target[0:]
```

## Preparing the Data into Training and Test Dataset
**Prepation for Logistic Regression Input**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, train_size = 0.8)
```

```python
len(X_train)
```

```python
len(X_test)
```

## Creating Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 10000)
```

```python
model.fit(X_train, y_train)
```

## Measuring Accuracy of the Model

```python
model.score(X_test, y_test)
```

### Time for Some Prediction!

**Selecting a Random Number**

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
import numpy as np
model2Darray = np.array(digits.data[67]).reshape(1, -1)
```

```python
prediction = model.predict(model2Darray)
print(prediction)
```

**Checking where Matrix did wrong, how it is performing: Enter Confusion Matrix**

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)
cm
```

**Visualizing the Confusion Matrix with a Heat Map**

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

**The Heat Map is a great way to look at a glance when the model*** ***'recognized'*** ***the actual digits as they are, when it did not recognize the right number, and what number the model termed it to be in that case.**

***For Example:*** **The Heat Map shows 6 was correctly recognized 30 times, but 2 times although the number was 6, the model guessed it to be 2.**
